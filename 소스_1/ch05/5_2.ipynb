```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
y = 1 / (1+np.exp(-x))

plt.plot(x,y)
plt.plot(x,y*(1-y))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10,10,0.1)
H = np.tanh(x)

plt.plot(x,H)
plt.plot(x,1-H**2)
plt.show()
```
